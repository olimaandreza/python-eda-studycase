# Objetivo

Realizar análise exploratória em uma base de dados da National Health and Nutrition Examination Survey (NHANES), que é uma pesquisa anual conduzida pelo National Center for Health Statistics (NCHS) do Centro de Controle e Prevenção de Doenças (Centers for Disease Control - CDC) dos Estados Unidos para avaliar a saúde e nutrição de adultos e crianças dos Estados Unidos.

A análise consiste em avaliar duas bases de dados:
1. DEMO_PHQ: Contém o perfil de indivíduos (adultos maiores de 18 anos) com sintomas depressivos nos EUA no período de 2005-2006
2. PAG_HEI: Contém informações sobre hábitos saudáveis de alimentação e atividade física 

Perguntas a serem respondidas: 

(1) Qual o perfil de indivíduos (adultos maiores de 18 anos) com sintomas depressivos nos EUA no período de 2005-2006?

(2) Hábitos saudáveis de alimentação e atividade física estão associados a menores índices de depressão nesta população?

## Por que realizar EDA?

Uma boa EDA fornece conhecimento suficiente para criar uma intuição sobre o dataset, revela quais são as features mais relevantes e as que carregam menos informações, evidencia uma noção geral sobre as variáveis numéricas e categóricas, identifica padrões, relações, distribuições, dados faltantes e outliers.

# Extra - Testes de Hipótese

Para realmente definir se os dados visualmente apresentam alguma discrepância, faz-se necessário realizar testes estatísticos. Serão tratados apenas alguns exemplos, existe uma variedade de testes aplicáveis que valem a pena um estudo mais aprofundado.

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from scipy.stats import f_oneway
```

```python
# ANOVA DPQ vs. Alimentação saudável: 
for i in df_category.columns[0:9]:

    db_aux = df[[i, "HEI2015_TOTAL_SCORE"]].dropna()

    stat, p = f_oneway(db_aux[(db_aux[i] == '0')]["HEI2015_TOTAL_SCORE"],
                    db_aux[(db_aux[i] == '1')]["HEI2015_TOTAL_SCORE"],
                    db_aux[(db_aux[i] == '2')]["HEI2015_TOTAL_SCORE"],
                    db_aux[(db_aux[i] == '3')]["HEI2015_TOTAL_SCORE"])
    print(i)
    print('stat=%.3f, p=%.3f' % (stat, p))
```

 Neste teste, testa-se a hipótese nula de que "não existe diferença entre os níveis de depressão com relação à alimentação saudável" e a hipótese alternativa de que existe pelo menos uma diferença. Ao nível de significância de 5%, apenas DPQ010, DPQ080 e DPQ090 não apresentaram evidências para rejeitar a hipótese nula, mesmo que visivelmente não tenha sido identificado este padrão nas demais variáveis.

```python
# ANOVA DPQ vs. exercícios físicos: 
for i in df_category.columns[0:9]:

    db_aux = df[[i, "PAG_MINW"]].dropna()

    stat, p = f_oneway(db_aux[(db_aux[i] == '0')]["PAG_MINW"],
                    db_aux[(db_aux[i] == '1')]["PAG_MINW"],
                    db_aux[(db_aux[i] == '2')]["PAG_MINW"],
                    db_aux[(db_aux[i] == '3')]["PAG_MINW"])
    print(i)
    print('stat=%.3f, p=%.3f' % (stat, p))
```

 Neste segundo teste, testa-se a hipótese nula de que "não existe diferença entre os níveis de depressão com relação à prática de exercícios físicos" e a hipótese alternativa de que existe pelo menos uma diferença. Ao nível de significância de 5%, apenas  DPQ080 e DPQ090 não apresentaram evidências para rejeitar a hipótese nula. Note-se que DPQ080, pelo box plot, aparenta ter diferença significativa, porém, como a variação dos dados é muito grande ou não existem informações suficientes, o teste não detecta essa diferença.

```python
db_aux = df[["RIAGENDR", "DPQ_SCORE"]].dropna()


tukey = pairwise_tukeyhsd(db_aux['DPQ_SCORE'],
                  db_aux['RIAGENDR'],
                  alpha = 0.05)

print(tukey)
```
